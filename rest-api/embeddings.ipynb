# Embeddings API

This notebook demonstrates the use of the TrustGraph embeddings API to compute embeddings from input text.

```python
import requests
import json
```

```python
# ID of flow
flow = "default"
base_url = "http://localhost:8088"
```

```python
# URL of the TrustGraph embeddings API
url = f"{base_url}/api/v1/flow/{flow}/service/embeddings"
```

```python
# The input consists of a text block
input = {
    "text": "What is the highest risk aspect of running a space shuttle program?  Provide 5 detailed reasons to justify our answer.",
}
```

```python
# Invoke the API, input is passed as JSON
resp = requests.post(url, json=input)
```

```python
# Should be a 200 status code
resp.status_code
```

```python
# Parse the response as JSON
object = resp.json()
```

```python
# The response is an array of vectors.  The number of embeddings depends on the embedding model in use
object["vectors"]
```
